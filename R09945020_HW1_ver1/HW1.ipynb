```python
import numpy as np
from matplotlib import pyplot 
from PIL import Image
```

```python
image=Image.open('lena.bmp')
```

```python
X=np.array(image)
```

```python
X.shape
```

```python
pyplot.imshow(X,cmap='gray')
```

```python
Y=np.zeros(X.shape)
```

```python
for i in range(X.shape[0]):
    Y[i]=X[X.shape[0]-1-i]
```

```python
pyplot.imshow(Y,cmap='gray')
```

```python
Z=np.zeros(X.shape)
```

```python
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Z[i][j]=X[i][X.shape[1]-1-j]
```

```python
pyplot.imshow(Z,cmap='gray')
```

```python
H=np.zeros(X.shape)
```

```python
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        H[i][j]=X[X.shape[0]-1-i][X.shape[1]-1-j]
```

```python
pyplot.imshow(H,cmap='gray')
```

```python
rotated = image.rotate(-45)
```

```python
R=np.array(rotated)
```

```python
pyplot.imshow(R,cmap='gray')
```
